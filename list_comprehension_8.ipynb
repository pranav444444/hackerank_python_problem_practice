```python
x=1
y=1
z=2
n=3
```

```python
for i in range(x+1):
  print('this is x: ',i)
  for j in range(y+1):
    print('this is y: ',j)
    for k in range(z+1):
      print('this is z: ',k)
```

```python
for i in range(x+1):
  for j in range(y+1):
    for k in range(z+1):
        print(i,j,k)
```

```python
for i in range(x+1):
  for j in range(y+1):
    for k in range(z+1):
      if (i+j+k) != n:
        print(i,j,k)
```

```python
#now solve this probelm by list comprehensiom
```

```python
new_list=[[i,j,k] for i in range(x+1) for j in range(y+1) for k in range(z+1) if (i+j+k) != n]
```

```python
print(new_list)
```
